# Data Preprocessing

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

raw_data = pd.read_csv("Results_21Mar2022.csv")
raw_data.head()
```

```python
raw_data.info()
```

```python
data = raw_data.iloc[:, 2:]

data.head()
```

```python
data.describe()
```

```python
grouped_data = data.groupby(["sex", "age_group", "diet_group"]).mean().reset_index()
grouped_data.head()
```

```python
grouped_data.info()
```

```python
grouped_data.describe()
```

```python
sex_grouped_data = (
    grouped_data.iloc[:, 1:12].groupby(["age_group", "diet_group"]).mean().reset_index()
)
sex_grouped_data
```

```python
age_sex_grouped_data = (
    sex_grouped_data.iloc[:, 1:11].groupby(["diet_group"]).mean().reset_index()
)
age_sex_grouped_data
```

```python
# Define the new order of the diet groups
order = {"vegan": 0, "veggie": 1, "fish": 2, "meat50": 3, "meat": 4, "meat100": 5}

# order the diet groups with the new order
df = age_sex_grouped_data
df["diet_group"] = pd.Categorical(
    df["diet_group"], categories=order.keys(), ordered=True
)

df = df.sort_values("diet_group")
df
```

```python
df.columns = [
    "diet_group",
    "GHGs(CO2)",
    "Land Use",
    "Water Scarcity",
    "Eutrophication",
    "CH4 Emissions",
    "N2O Emissions",
    "Biodiversity",
    "Water Use",
    "Acidification",
]
df
```

# Draw the Radar Chart

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# the data to plot
plot_data = df

# normalise the data
data_normalised = (
    plot_data.set_index("diet_group") - plot_data.set_index("diet_group").min()
) / (plot_data.set_index("diet_group").max() - plot_data.set_index("diet_group").min())
data_normalised = data_normalised.reset_index()

# number of variable
labels = np.array(plot_data.columns[1:])
num_vars = len(labels)

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

fig, ax = plt.subplots(figsize=(8, 8), subplot_kw=dict(polar=True))
# colors = ["green", "blue", "cyan", "yellow", "red", "purple", "pink"]
# colors = ["#FD6D5A", "#FEB40B", "#6DC354", "#994487", "#518CD8", "#443295"]

for idx, row in data_normalised.iterrows():
    values = row.drop("diet_group").values.flatten().tolist()
    values += values[:1]
    ax.fill(angles, values, alpha=0.1)
    # ax.fill(angles, values, color=colors[idx], alpha=0.05)
    # print(idx, colors[idx])
    ax.plot(
        angles,
        values,
        marker="o",
        markersize=3,
        linewidth=1,
        label=row["diet_group"],
        # color=colors[idx]
    )

ax.set_thetagrids(np.degrees(angles[:-1]), labels)
plt.legend(loc="upper right", bbox_to_anchor=(0.1, 0.1))
plt.show()
```
